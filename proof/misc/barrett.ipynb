```python
import matplotlib.pyplot as plt

k = 8
kk = 2 * k
n = 2 ** k
n2 = 2 ** kk
eta = 13
q = n * eta + 1
r = 2 ** 15

f = lambda a : a - ((a * ((n2 // q) + 1)) // n2) * q
x = [a for a in range(- r , r)]
y = [f(a) for a in x]
down = [0 for a in x]
up = [2 * q for a in x]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.plot(x,down)
ax.plot(x,up)
plt.show()

# f is affine by parts
```

```python
import math
v = (n2 // q) + 1
m = lambda l : math.ceil ((n2*l)/v)

argmins_labels = [l for l in range(1,(v*r)//n2)]
argmins = [m(l) for l in argmins_labels]
mins = [f(a) for a in argmins]

fig, ax = plt.subplots()
ax.plot(x,y)
ax.plot(x,down)
ax.plot(x,up)
ax.plot(argmins,mins,'.')
plt.show()

# f's lowest inflexion points are given by m
```
